### 1) Import des bibliotheques

```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
```

### 2) chargement du csv

```python
df = pd.read_csv('../data/online_retail_II.csv')	
df
```

### 3) Exploration

```python
df.info()
df.shape
```

```python
df.isnull().sum()
```

#### 3-1) Suppression valeurs nuls

```python
df = df.dropna(subset=['Customer ID'])
df.isnull().sum()
```

#### 3-2) Conversion des types

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Customer ID'] = df['Customer ID'].astype(int)
```

#### 3-3) Suppréssion Quantités et prix négatifs ou nuls

```python
df = df[df['Quantity'] > 0]
df = df[df['Price'] > 0]
```

#### 3-4) Suppréssion des Annulations : Invoice commençant par 'C'

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

### 4) Création colonnes revenu

```python
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
# Resume apres netoyage : 
df.info()
df.shape
```

```python
# Sauvegarde new df
df.to_csv("data/cleaned_online_retail.csv", index=False)
```
