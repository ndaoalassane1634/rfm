### 1) Import des librairies et chargement du fichiers 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/cleaned_online_retail.csv')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

### 2) Date de référence

```python
analysis_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
analysis_date
```

### 3) Agregation par types de clients

```python
rfm = df.groupby('Customer ID').agg(
    Recency=('InvoiceDate', lambda x: (analysis_date - x.max()).days),
    Frequency=('Invoice', 'nunique'),
    Monetary=('Revenue', 'sum')
).reset_index()

rfm
```

```python
rfm.info()
```

### 4) Attribution des scores par quintiles

```python
# Récence : score 5 = achat récent (faible nb de jours)

# Fréquence : score 5 = achats très fréquents

# Monétaire : score 5 = gros dépensiers
```

```python
rfm['R'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1]).astype(int)
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
rfm['M'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5]).astype(int)
```

```python
# Score global
rfm['RFM_Score'] = rfm[['R', 'F', 'M']].sum(axis=1)
rfm
```
